## Anova oneway test

```python
from scipy.stats import *
val1 = df['LotArea'].head(20)
val2 = df['SalePrice'].head(20)
f_statistics,p_value = f_oneway(val1,val2)

print("Statistics :",f_statistics)
print("P_value :",p_value)

alpha = 0.05
if p_value > alpha:
    print("Rejected the null hypothesis.")
else:
    print("Accepted the null hypothesis.")
```

## Z_Score:

```python
mean = np.mean(dff['SalePrice'])
std_dev = np.std(dff['SalePrice'])
z_score = (dff['SalePrice']- mean)/std_dev

print("Mean :", mean)
print("Std_)dev :", std_dev)
print("z_core value :",z_score)
```
